## Basic and Advanced operations using pandas

```python
import pandas as pd
```

```python
# Loading the apple stock data dataset
dataset=pd.read_csv('C:/Users/varshini y r/OneDrive/Desktop/dva1/tiktok_performance.csv',index_col=0)
print(dataset)
```

```python
dataset.shape
```

```python
# Indexing
## index the row with index 101
dataset.loc[101]
```

```python
## index the last row
dataset.iloc[-1]
```

```python
# Slicing
## slice of the first 10 rows
dataset.iloc[0:10]
```

```python
# Iterating
for index,row in dataset.iterrows():
    print(index,row)
```

```python
# Filtering
dataset.filter(items=["User_ID","Username","Views",]).head()
```

```python
# Sorting
dataset.sort_values(by=["User_Followers"])[["User_Followers"]].head(5)
```

```python
# Reshaping
pivot=dataset.pivot(index='User_ID',columns='Hashtags',values='Views')
print(pivot.head())
```
