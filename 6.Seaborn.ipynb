# Adavantages of Seaborn
## * Simplifies complex visualizations with less code and fewer lines.
## * Built specifically for statistical plotting, making it easier to create plots like heatmaps, pair plots, and categorical plots.
## * Comes with several pre-defined themes and color palettes that enhance the visual appeal of plots.
## * Directly accepts pandas DataFrames for plotting, simplifying data manipulation.
## * Provides a good balance between simplicity and customization, allowing easy adjustments to aesthetics.

# Controlling Figure Aesthetics
## Matplotlib is highly customizable. But it also has the effect that it is very inconvenient, as it can take a 
long time to adjust all necessary parameters to get your desired visualization. In Seaborn, we can use customized themes and a highlevel interface for controlling the appearance of Matplotlib figures.

```python
# The following code snippet creates a simple line plot in Matplotlib:
import matplotlib.pyplot as plt
plt.figure()
x1 = [10, 20, 5, 40, 8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label='Group A')
plt.plot(x2, label='Group B')
plt.legend()
plt.show()
```

```python
# To switch to the Seaborn defaults, simply call the set() function:
import seaborn as sns
sns.set()
plt.figure()
x1 = [10, 20, 5, 40, 8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label='Group A')
plt.plot(x2, label='Group B')
plt.legend()
plt.show()
```
