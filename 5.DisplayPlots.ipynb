# Basic Plots - Histogram, Box Plot, Scatter Plot, Bubble Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Histogram
data=np.random.normal(loc=50,scale=10,size=1000)
plt.figure(figsize=(10,6))
plt.hist(data,bins=30,color='lightgreen',alpha=0.7,edgecolor='black')
plt.title('Histogram of randomly generated data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y',linestyle='--')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(10,6))
plt.boxplot(data,vert=True,patch_artist=True,boxprops=dict(facecolor='indigo',color='blue'),medianprops=dict(color='red'),whiskerprops=dict(color='blue'),capprops=dict(color='blue'))
plt.title('Box plot of randomly generated data')
plt.ylabel('Values')
plt.show()
```

```python
# Scatter plot
np.random.seed(0)
x=np.random.rand(100)*100
y=np.random.rand(100)*100
plt.figure(figsize=(10,6))
plt.scatter(x,y,color='blue',alpha=0.6,edgecolors='black',s=100)
plt.title('Scatter plot of Randomly generated data')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.show()
```

```python
# Bubble Plot
np.random.seed(0)
x=np.random.rand(50)*100
y=np.random.rand(50)*100
sizes=np.random.rand(50)*1000
plt.figure(figsize=(8,6))
bubble=plt.scatter(x,y,s=sizes,alpha=0.5,c='lightblue',edgecolors='blue',linewidth=2)
plt.title('Bubble plot of randomly generated data')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.show()
```
