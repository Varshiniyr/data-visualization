# Pyplot Basics

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the Agriculture crop yield dataset
dataset=pd.read_csv('C:/Users/varshini y r/OneDrive/Desktop/dva1/crop_yield.csv')
df=dataset.head()
print(df)
```

```python
# Creating Figures
plt.figure(figsize=(12,6))

# Plotting crop yield
plt.bar(df['Crop'],df['Yield'],color='skyblue',alpha=0.7,label='Yield')

# Formatting strings in the title and axes
plt.title('Agricultural Crop Yield',fontsize=16)
plt.xlabel('Crops',fontsize=14)
plt.ylabel('Yield',fontsize=14)

# Adding text to the bars
for index,value in enumerate(df['Yield']):
    plt.text(index,value+0.1,str(value),ha='center',fontsize=12)

# Adding a legend
plt.legend()

# Adding annotations
plt.annotate('Highest Yield',xy=(3,6.0),xytext=(3,6.5),arrowprops=dict(facecolor='black',shrink=0.05))

# Displaying the figure
plt.grid(axis='y',linestyle='--')
plt.tight_layout()
plt.show()

# Saving the figure
plt.savefig('crop_yield.png')
```
